## Bar Chart

* Bivariate Analysis
* Numerical vs Categorical
* Use case - Aggregate analysis of groups
 

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow', 'pink']

plt.bar(colors, children)
```

```python
# color
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow', 'pink']

plt.bar(colors, children, color = 'red')
```

```python
# labelling
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow', 'pink']

plt.bar(colors, children)
plt.xlabel('colors')
plt.ylabel('Number of children')
plt.title('Favorite color of Children')
```

```python
# horizontal bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow', 'pink']

plt.barh(colors, children)
```

```python
# Multiple bar charts
df = pd.read_csv(fr"C:\Users\sanje\Desktop\Python_Frameworks_For_Machine_Learning\datafiles\matplotlib_basics\batsman_season_record.csv")
df.head()
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'], width = 0.2)
plt.bar(np.arange(df.shape[0]),df['2016'], width = 0.2)
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'], width = 0.2)
```

```python
#plt.xticks()
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'], width = 0.2, color = 'yellow')
plt.bar(np.arange(df.shape[0]),df['2016'], width = 0.2, color = 'red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'], width = 0.2 , color ='blue')

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
# problem -> Overlapping due to long name of category
children = [10,20,40,10,30]
colors = ['red red red red red red red red','blue','green green green','yellow yellow yellow', 'pink']

plt.bar(colors, children)
```

```python
# solution
children = [10,20,40,10,30]
colors = ['red red red red red red red red','blue','green green green','yellow yellow yellow', 'pink']

plt.bar(colors, children)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Stacked Bar Chart
plt.bar(df['batsman'], df['2015'], label = '2015')
plt.bar(df['batsman'], df['2016'], bottom = df['2015'], label = '2016')
plt.bar(df['batsman'], df['2017'], bottom = (df['2015'] + df['2016']), label = '2017')
plt.legend()
plt.show()
```

## Pie Chart

* Univariate/Bivariate Analysis
* Categorical vs Numerical
* Use case - To find contribution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subject_marks = ['english','science','maths','social_studies','nepali']
plt.pie(data, labels = subject_marks)
plt.show()
```

```python
# dataset
df = pd.read_csv(fr"C:\Users\sanje\Desktop\Python_Frameworks_For_Machine_Learning\datafiles\matplotlib_basics\gayle_175.csv")
df
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels =df['batsman'],autopct='%0.2f%%', colors =['blue','green','yellow','pink','brown','purple'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels =df['batsman'],autopct='%0.2f%%', explode =[0.1,0,0.3,0.5,0,0])
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels =df['batsman'],autopct='%0.2f%%', explode =[0.1,0,0.3,0.5,0,0], shadow=True)
plt.show()
```
