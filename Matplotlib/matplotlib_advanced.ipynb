## Annotations

## Subplots

```python
# A diff way to plot a graphs
batters.head()
```

```python
fig , axes = plt.subplots(figsize = (15,6))

axes.scatter(batters['avg'], batters['strike_rate'], color = 'red', marker = '+')
axes.set_title('Average vs Strike Rate')
axes.set_xlabel('Average')
axes.set_ylabel('Strike Rate')

fig.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols =1, sharex =True, figsize = (10,8))

axes[0].scatter(batters['avg'], batters['strike_rate'], color = 'red')
axes[1].scatter(batters['avg'], batters['runs'], color = 'green')

axes[0].set_title('Average Vs Strike Rate')
axes[0].set_ylabel('Strike Rate')

axes[1].set_title('Average Vs Runs')
axes[1].set_ylabel('Runs')
axes[1].set_xlabel('Average')
```

```python
#method1
fig, axes = plt.subplots(nrows=2, ncols =2, figsize = (10,10))

axes[0,0].scatter(batters['avg'], batters['strike_rate'], color = 'red')
axes[0,1].scatter(batters['avg'], batters['runs'], color = 'green')
axes[1,0].hist(batters['avg'], color ='purple')
axes[1,1].hist(batters['runs'], color ='yellow')

fig.show()
```

```python
#method2
fig = plt.figure()

ax1 = fig.add_subplot(2,1,1)
ax1.scatter(batters['avg'], batters['strike_rate'], color = 'red')
ax2 = fig.add_subplot(2,1,2)
ax2.scatter(batters['avg'], batters['runs'], color = 'green')
```

## 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
```

```python
xx.shape
```

```python
yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(15,10))

ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z, cmap = 'viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) +np.cos(yy)
fig = plt.figure(figsize=(15,10))

ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z, cmap = 'viridis')
fig.colorbar(p)
```

## Contour Plots

```python
z = xx**2 + yy**2
fig = plt.figure(figsize=(15,10))

ax = plt.subplot()
p = ax.contour(xx,yy,z, cmap = 'viridis')
fig.colorbar(p)
```

## Contour Fill

```python
z = xx**2 + yy**2
fig = plt.figure(figsize=(15,10))

ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap = 'viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) +np.cos(yy)
fig = plt.figure(figsize=(15,10))

ax = plt.subplot()
p = ax.contourf(xx,yy,z, cmap = 'viridis')
fig.colorbar(p)
```
