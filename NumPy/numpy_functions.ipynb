np.sort\
Return a sorted copy of an array.

```python
a = np.random.randint(low=1,high=100,size=15)
a
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
b
```

```python
np.sort(a)  #sort in ascending order by default
```

```python
np.sort(a)[::-1]  #sort in descending order
```

```python
np.sort(b)   #sort the elements of each row
```

```python
np.sort(b, axis=0) #sort the elements of each row
```

np.concatenate\
The np.concatenate() function concatenate a sequence of arrays along an existing axis.

```python
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)

print(c)
print(d)
```

```python
np.concatenate((c,d),axis=0)
```

```python
np.concatenate((c,d),axis=1)
```

np.where\
The np.where() function returns the indices of elements in an input array where the given condition is satisfied.

```python
a
```

```python
# find all indices with value greater than 50
np.where(a>50)
```

```python
# replace all values greater than 50 with 0
np.where(a>50,0,a)
```

```python
# replace all even numbers with 0
np.where(a % 2 == 0, 0, a)
```

Set Functions
   * np.union1d
   * np.intersect1d
   * np.setdiff1d
   * np.setxor1d
   * np.in1d

```python
m = np.array([1,2,3,4,5])
n = np.array([3,4,5,6,7])
```

```python
np.union1d(m,n)
```

```python
np.intersect1d(m,n)
```

```python
np.setdiff1d(m,n)
```

```python
np.setxor1d(m,n)   #removes the common elements
```

```python
np.in1d(m,1)
```

```python
m[np.in1d(m,1)]
```

```python
np.in1d(m,20)
```
