```python
# import titanic dataset
df = pd.read_csv(fr"c:\Users\sanje\Desktop\Python_Frameworks_For_Machine_Learning\datafiles\titanic.csv")
df.head(3)
```

```python
df['Name']
```

```python
# Common Functions
#lower
df['Name'].str.lower()
```

```python
# upper
df['Name'].str.upper()
```

```python
# capitalize
df['Name'].str.capitalize()
```

```python
# title
df['Name'].str.title()
```

```python
# len
df['Name'].str.len()
```

```python
# find the name of the passenger with the longest name
df['Name'][df['Name'].str.len() == df['Name'].str.len().max()].values[0]
```

```python
# strip   # remove leading and trailing spaces
df['Name'].str.strip()
```

```python
# split -> get
df['lastname'] = df['Name'].str.split(',').str.get(0)
```

```python
df.head(3)
```

```python
df[['title','firstname']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n = 1, expand=True)
```

```python
df.head(3)
```

```python
df['title'].value_counts()
```

```python
# replace
df['title'] = df['title'].str.replace('Ms.','Miss.')
df['title'] = df['title'].str.replace('Mlle.','Miss.')
```

```python
df['title'].value_counts()
```

```python
# filtering
# startswith/endswith
df[df['firstname'].str.startswith('A')]
```

```python
df[df['firstname'].str.endswith('A')]
```

```python
# isdigit/isalpha

df[df['firstname'].str.isdigit()]
```

```python
df[df['firstname'].str.isalpha()]
```

```python
# applying regex
# contains
#search john -> both case (upper and lower)
df[df['firstname'].str.contains('john', case=False)].head(3)
```

```python
# find lastnames with start and end char vowel
df[df['lastname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')].head(3)
```
