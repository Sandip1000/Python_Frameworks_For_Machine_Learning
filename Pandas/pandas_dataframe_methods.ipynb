```python
points_df =pd.DataFrame(
    {
        '1st point' :[(3,4),(6,-5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)
points_df
```

```python
def euclidean_distance(row):
    pt_A = row['1st point']
    pt_B = row['2nd point']
    return ((pt_A[0]-pt_B[0])**2 +(pt_A[1]-pt_B[1])**2)**0.5
```

```python
points_df['euclidean_distance'] = points_df.apply(euclidean_distance, axis =1)
points_df
```

```python
# corr
# Create a sample DataFrame
data = {
    'Age': [23, 25, 30, 45, 18],
    'Salary': [45000, 50000, 62000, 110000, 22000],
    'Hours_Worked': [40, 42, 50, 38, 20]
}
df = pd.DataFrame(data)

# 1. Default Pearson Correlation Matrix
pearson_matrix = df.corr()
print(pearson_matrix)

# 2. Spearman Rank Correlation (For non-linear relationships)
spearman_matrix = df.corr(method='spearman')
```

```python
# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pearson_matrix, annot=True, cmap='Blues', fmt='.2f', vmin=-1, vmax=1)

# Show the plot
plt.title('Correlation Heatmap')
plt.show()
```

```python
# nlargest and nsmallest(series and dataframe)
data = {
    'Country': ['South Africa', 'Canada', 'Australia', 'Maldives', 'Iceland'],
    'Population': [58812491, 37522972, 339757, 339757, 339757],
    'GDP': [203359, 183359, 153594, 14520, 10128]
}
df = pd.DataFrame(data)

# Get the 2 rows with the largest 'Population'
top_pop = df.nlargest(2, 'Population')
print(top_pop)
```

```python
# 'keep=all' returns all 3 matching rows for 339757, extending the output
all_ties = df.nsmallest(3, 'Population', keep='all')
print(all_ties)
```

```python
# Sorts by smallest population first, then breaks ties using lowest GDP
multi_sort = df.nsmallest(3, ['Population', 'GDP'])
print(multi_sort)
```

```python
# copy(series+dataframe)

df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})

# Create an independent copy
df_copy = df.copy()

# Modifying the copy does not alter 'df'
df_copy.at[0, 'A'] = 99
```
