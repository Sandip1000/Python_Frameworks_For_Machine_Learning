```python
# agg method
# passing dict
# soecify the aggregate functions applied to each column
movies.groupby('Genre').agg(
    {
        'Runtime': 'mean',
        "IMDB_Rating" : 'mean',
        "No_of_Votes" : 'sum',
        'Gross': 'sum',
        "Metascore" : 'min'
    }
)
```

```python
# passing list
movies.groupby("Genre")[['IMDB_Rating', 'No_of_Votes', 'Runtime','Gross', 'Metascore']].agg(['mean','max','min'])
```

```python
#combing both
movies.groupby('Genre').agg(
    {
        'Runtime': ['min','mean','max'],
        "IMDB_Rating" : ['mean','max'],
        "No_of_Votes" : 'sum',
        'Gross': 'sum',
        "Metascore" : 'min'
    }
)
```

```python
# split(apply) combine
# apply -> builtin function

movies.groupby('Genre')['IMDB_Rating'].apply(np.max)
```

```python
# find the number of movies starting with A for each genre

def temp(group):
     return group["Series_Title"].str.startswith('A').sum()

movies.groupby('Genre').apply(temp, include_groups = False)
```

```python
# find ranking of each movie in the group according to the IMDB rating


def rank_movie(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending= False)
    return group
 
movies.groupby('Genre').apply(rank_movie, include_groups = False)
```

```python
# find normalized IMDB rating groupwise

def normalize(group):
    group['norm_rating'] = (group['IMDB_Rating']- group['IMDB_Rating'].min())/(group['IMDB_Rating'].max()- group['IMDB_Rating'].min())
    return group
```

```python
movies.groupby('Genre').apply(normalize, include_groups = False)
```

```python
# groupby on multiple cols
duo = movies.groupby(['Director','Star1'])
duo
# size
duo.size()
# get_group
duo.get_group(('Aamir Khan', 'Amole Gupte'))
```

```python
# find the most earning actor -> director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# find the best(in-terms of metascore(average)) actor -> genre combo
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore', ascending = False).head(1)
```

```python
# agg on multiple groupby
duo[['IMDB_Rating','Runtime']].agg(['min','max','mean'])
```

```python
ipl = pd.read_csv(fr"C:\Users\sanje\Desktop\Python_Frameworks_For_Machine_Learning\datafiles\deliveries.csv")
ipl.head(2)
```

```python
ipl.shape
```

```python
 # find the top 10 batsman in terms of runs scored
ipl.groupby("batsman")['batsman_runs'].sum().sort_values(ascending= False).head(10)
```

```python
# Find the batsman  with max number of sixes
six = ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1)
```

```python
# find batsman with most number of 4's and 6's in last five overs
temp_df = ipl[ipl['over'] > 15]
temp_df = temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6)]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find V Kohli's record against all teams

temp_df = ipl[ipl['batsman'] == 'V Kohli']
temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Create a function that can return the highest score of any batsman

def highest_run(batsman):
    temp_df = ipl[ipl['batsman'] == batsman]
    return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0].item()

highest_run('V Kohli')
```
