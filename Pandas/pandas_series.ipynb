## What is Pandas?

Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.

A pandas series is like a column in a table. It is a 1-D array holding data of any type.

## Importing Pandas

```python
import pandas as pd
import numpy as np
```

## Series Using Lists

```python
# string (object)
country = ['Nepal', 'India', 'Bangladesh', 'China','Srilanka']

pd.Series(country)
```

```python
# integers
runs = [13,24,56,78,100]

runs_series = pd.Series(runs)
runs_series
```

```python
# custom index

marks = [67,57,89,100]
subjects = ['maths','english','science','nepali']

pd.Series(marks, index=subjects)
```

```python
# setting a name

marks_series = pd.Series(marks, index=subjects, name="marks of sandip" )
marks_series
```

## Series From Dictionary

```python
marks = {   
    'maths': 67,
    'english':57,
    'science':89,
    'nepali':100
}

marks_series = pd.Series(marks,name="marks of sandip")
marks_series
```
