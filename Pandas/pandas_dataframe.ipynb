```python
# Write A Function That Can Return The Track Record Of 2 Teams Against Each Other
def trackRecord(df :pd.DataFrame, Team1 :str, Team2 :str) -> pd.DataFrame:
    mask = ((df['Team1'] == Team1) & (df['Team2'] == Team2)) | ((df['Team1'] == Team2) & (df['Team2'] == Team1))
    Matches_Won_By_Team1 = df[mask][df[mask]['WinningTeam'] == Team1].shape[0]
    Matches_Won_By_Team2 = df[mask][df[mask]['WinningTeam'] == Team2].shape[0]
    print(f"Total Match Played: {df[mask].shape[0]}")
    print(f"{Team1} Won : {Matches_Won_By_Team1}")
    print(f"{Team2} Won : {Matches_Won_By_Team2}")


trackRecord(ipl, 'Rajasthan Royals', 'Mumbai Indians')
```
