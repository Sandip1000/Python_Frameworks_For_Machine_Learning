```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Why Seaborn?

* provides a layer of abstraction hence simpler to use
* better asthetics
* more graphics included

Types of Functions
* Figure Level
* Axis Level

Main Classification
* Relational Plot -> Scatterplot, Lineplot
* Distribution Plot -> Histogram, KDE Plot, Rug Plot
* Categorical Plot  -> Bar Plot, Count Plot, Box Plot, Violin Plot, Swarm Plot

## 1. Relational Plot

* To see the statistical relation between 2 or more variables
* Bivariate Analysis
  
Plots under this section
* Scatter Plot
* Line Plot

```python
# scatter plot  -> axis level function
sns.scatterplot(data=tips, x ='total_bill', y="tip", hue ='sex', style='time', size = 'size')
```

```python
# Mostly used and often preffered
# sns.replot() -> Figure level function  -> square shape default
sns.relplot(data=tips, x='total_bill', y='tip', kind ='scatter', hue='sex', style='time', size='size')
# This figure overall represents the information in five dimension
```

```python
# Line plot
gap = px.data.gapminder()
gap.sample(5)
```

```python
temp_df = gap[gap['country'] == 'Nepal']
temp_df
```

```python
# axis level
sns.lineplot(data = temp_df, x ='year',y ='lifeExp')
```

```python
# using relplot  -> figure level
sns.relplot(data = temp_df, x ='year', y ='lifeExp', kind ='line')
```

```python
temp_df = gap[gap['country'].isin(['India','China', 'Nepal','Brazil','Germany'])]
temp_df.sample(5)
```

```python
sns.relplot(kind ='line', data = temp_df, x ='year', y ='lifeExp', hue ='country', style='continent')
```

```python
sns.lineplot(data = temp_df, x ='year', y ='lifeExp', hue ='country', size='continent')
```

```python
# facet plot -> only with figure level function  not with scatterplot and lineplot function(axis level)
sns.relplot(data=tips, x='total_bill', y='tip',kind ='scatter', col='sex', hue='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip',kind ='scatter', col='smoker', row ='sex')
```

```python
# col wrap
sns.relplot(kind = 'scatter', data = gap, x ='lifeExp', y ='gdpPercap', col = 'year', col_wrap=3)
```

## 2. Distribution Plots

* used for univariate analysis
* used to find out the distribution
* Range of the observation
* Central Tendency
* Is the data bimodal?
* Are there outliers?

Plots under distribution plot
* histplot
* kdeplot
* rugplot

```python
# figure level -> displot
# axis level -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram
sns.histplot(data = tips, x ='total_bill')
```

```python
sns.displot(data = tips, x = 'total_bill', kind = 'hist')
```

```python
# bins parameter
sns.displot(data = tips, x = 'total_bill', kind = 'hist', bins = 20)
```

```python
# It is also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical variables.

# countplot
sns.displot(data = tips, x = 'day', kind ='hist')
```

```python
# hue parameter
sns.displot(data = tips, x = 'tip', kind = 'hist', hue='sex')
```

```python
# element -> step
sns.displot(data = tips, x = 'tip', kind = 'hist', hue='sex', element = 'step')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.displot(data = titanic, x = 'age', kind = 'hist', element = 'step', hue ='sex', col='survived')
```

```python
# faceting using col and row -> will not work on axis level function like histplot()
sns.displot(data = tips, x = 'tip', kind = 'hist', col='sex', element = 'step')
```

```python
# kdeplot -> kernel density estimation plot
# Rather than using discrete bins, a KDE plot smooths the observations with a Guassian kernel, producing a continuous density estimate.
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(kind = 'kde', data=tips, x='total_bill')
```

```python
# hue -> fill
sns.displot(kind = 'kde', data=tips, x='total_bill',hue='sex')
```

```python
sns.displot(kind = 'kde', data=tips, x='total_bill',hue='sex', fill=True,height=10, aspect=1.5)
```

```python
# Rugplot
# Plot marginal distribution by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location of individual observations in an unobstrusive way.
sns.kdeplot(data = tips, x ='total_bill')
sns.rugplot(data = tips, x = 'total_bill')
```

```python
# Bivariate Histogram
# A bivariate histogram bins the data within rectangles that tile the plot.
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data = tips, x ='total_bill', y = 'tip')
```

```python

sns.displot(kind = 'hist', data = tips, x ='total_bill', y = 'tip')
```

```python
# Bivariate kdeplot
# A bivariate KDE plot smoothes the (x,y) observations with a 2D Gaussian

sns.kdeplot(data = tips, x ='total_bill', y = 'tip')
```

```python
sns.displot(kind = 'kde', data = tips, x ='total_bill', y = 'tip')
```

## 3. Matrix Plot

* Heatmap
* Clustermap

```python
# Heatmap
# Plot rectangular data as a color-encoded matrix
# heatmap() is axis level function
temp_df= gap.pivot(index = 'country',columns='year', values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot and linewidth
temp_df = gap[gap['continent'] =='Europe'].pivot(index = 'country',columns='year', values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df, annot=True, linewidth = 0.5)
```

```python
# cmap
temp_df = gap[gap['continent'] =='Europe'].pivot(index = 'country',columns='year', values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(temp_df, annot=True, linewidth = 0.5, cmap = 'Blues')
```

```python
# Clustermap
# Plot a matrix dataset as a hierarchically-clustered heatmap.
# This function requires scipy to be available.
iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
