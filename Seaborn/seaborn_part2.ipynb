## Box Plot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary(minimum, first quartile(Q1), median(Q2), third quartile(Q3), and maximum). It can tell us about outliers and what their values are. Boxplots can also tell us if our data is symmetrical, how tightly our data is grouped and if and how our data is skewed.

InterQuartile Range(IQR) = Q3 - Q1

Minimum = Q1 - 1.5 * IQR

Maximum = Q3 + 1.5 * IQR


```python
# Box Plot
sns.boxplot(data = tips, x= 'sex', y ='total_bill')
```

```python
# using catplot
sns.catplot(data = tips, x= 'sex', y ='total_bill', kind='box')
```

```python
# hue
sns.boxplot(data = tips, x= 'day', y ='total_bill', hue = 'sex')
```

```python
# single boxplot -> numerical column
sns.boxplot(data = tips, y = 'total_bill')
```

## Categorical Estimate Plot -> For Central Tendency

#### Barplot 

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

```python
# barplot
sns.barplot(data = tips, x = 'sex', y = 'total_bill')
```

```python
# without errorbar -> no confidence interval
sns.barplot(data = tips, x = 'sex', y = 'total_bill', errorbar=None)
```

```python
sns.catplot(data = tips, x = 'sex', y = 'total_bill', kind = 'bar')
```

```python
sns.catplot(data = tips, x = 'sex', y = 'total_bill', kind = 'bar', hue ='smoker')
```

```python
# estimator
sns.catplot(data = tips, x = 'sex', y = 'total_bill', kind = 'bar', estimator='std')
```

#### Countplot

A special case for the bar plot is when we want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative variable.

```python
# countplot
sns.countplot(data = tips, x ='sex')
```

```python
sns.countplot(data = tips, x ='sex', hue = 'day')
```

## Regression Plots
* regplot
* lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
# axes level
# hue parameter is not available
sns.regplot(data = tips, x = 'total_bill' , y = 'tip')
```

```python
# figure level
sns.lmplot(data = tips, x = 'total_bill' , y = 'tip')
```

```python
# hue parameter is available
sns.lmplot(data = tips, x = 'total_bill' , y = 'tip', hue ='sex')
```

```python
# residual plot or residplot -> plot the residual error
# residual error = prediction error
sns.residplot(data = tips, x = 'total_bill', y  = 'tip')
```

## Second Way to Plot Facet Plots -> FacetGrid

```python
# figure level -> relplot -> displot -> catplot -> lmplot
sns.catplot(data = tips, x = 'sex', y = 'total_bill', kind = 'violin' , col ='day',row = 'time')
```

```python
g  = sns.FacetGrid(data = tips, col ='day' , row = 'time', hue ='smoker')
g.map(sns.violinplot, 'sex', 'total_bill')
g.add_legend()
```

## Plotting Pairwise Relationship(PairGrid Vs Pairplot)

```python
sns.pairplot(data=iris)
```

```python
sns.pairplot(data = iris , hue = 'species')
```

```python
# pair grid
g = sns.PairGrid(data = iris, hue = 'species')
# g.map
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data = iris, hue = 'species', vars = ['sepal_width', 'petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
